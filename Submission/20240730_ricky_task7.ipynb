## 4. How can you find the customer who has spent the most based on the Total column in the table_transaction

```python
# Pakai cara di atas tapi kali ini diberi nama dataframenya
TotalSpent = df_join.groupby(['CustomerId', 'FirstName'])[['Total']].sum().reset_index()

# Cari yang tertinggi dari dataframe TotalSpent menggunakan max
HighestSpent = TotalSpent['Total'].max()

print(HighestSpent)
```

```python
# Kita buat dengan nama dataframe baru lalu dipanggil TotalSpent di sini agar nanti preview yang tampil rapi lengkap dg Customerid dan FirstName
HighestSpender = TotalSpent[TotalSpent['Total'] == HighestSpent]

print(HighestSpender)
```

## 5. How would you join the two tables to include transactions for only those customers who are in a specific country, such as 'USA'?

```python
# Table Preparation
col = ['CustomerId', 'Country', 'Total']
df_left = df_transaction[col]
df_right = df_customer_id[['FirstName', 'CustomerId']]

# KIta join by left dulu soalnya pakai Country di csv Transaction
df_join = df_left.join(df_right.set_index('CustomerId'), on='CustomerId')

# Di sini kita filter ke transaction USA only
df_usa = df_join[df_join['Country'] == 'USA']

# Display result
print(df_usa)
```

# 6. Write a command to count the number of transactions made by each customer.


```python
#Di sini kita bikin dulu dataframe baru yaitu transaction_counts, value_counts di sini beri counter bila CustomerId sama nanti akan dihitung +1 per 
#CustomerId yang akan ditambahkan ke Column TransactionCount

transaction_counts = df_transaction['CustomerId'].value_counts().reset_index()
transaction_counts.columns = ['CustomerId', 'TransactionCount']
```

```python
#ini kita bikin data frame df_result untuk dipanggil melalui print nanti

df_result = df_customer_id.join(transaction_counts.set_index('CustomerId'), on='CustomerId')
```

```python
#Lalu tinggal diprint df_result nya

print(df_result[['CustomerId', 'FirstName', 'TransactionCount']])
```

## 7. How can you list all customers along with the number of transactions and their total spending?

```python
# Pertama kita buat df baru yaitu transaction_summary utk nanti dipanggil, transaction_summary ini kita masukkan groupby pada df_transaction dimana kita
# menggambil CustomerId sebagai pusatnya, lalu kita copas TransactionCount yang tadi dibuat dan sertakan df baru namanya TotalSum

transaction_summary = df_transaction.groupby('CustomerId').agg(
    TransactionCount=('CustomerId', 'count'),
    TotalSum=('Total', 'sum')
).reset_index()

# ketitka df TransactionCount dan TotalSum sudah digabungkan dalam satu frame transaction_summary, kita tinggal masukan ke df_result utk dipanggil print
df_result = df_customer_id.join(transaction_summary.set_index('CustomerId'), on='CustomerId')

# Tinggal di print result nya
print(df_result[['CustomerId', 'FirstName', 'TransactionCount', 'TotalSum']])
```

## 10 How can you extract and display the top 5 cities with the highest total transaction amounts?

```python
# Table Preparation
col = ['CustomerId', 'Total', 'City']
df_left = df_transaction[col]
df_right = df_customer_id[['FirstName', 'CustomerId']]

# KIta join by left dulu soalnya pakai City di csv Transaction
df_join = df_left.join(df_right.set_index('CustomerId'), on='CustomerId')

# Di sini kita groupby City dan Total yang sudah di sum
total_bycity = df_join.groupby('City')['Total'].sum().reset_index()

# Di sini kita temukan 5 n terbesar dari dataframe total_bycity
top_5_cities = total_bycity.nlargest(5, 'Total')

# Display resultnya
print(top_5_cities)
```
