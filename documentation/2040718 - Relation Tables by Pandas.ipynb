# Left join

```python
df_customer_id.dtypes
```

```python
# Table Preparation
col = ['CustomerId ', 'BillingCity         ']
df_left = df_transaction[col]
df_right = df_customer_id[['FirstName ', 'CustomerId ']]
```

```python
df_join = pd.merge(df_left, df_right, how='left', on='CustomerId ')
```

```python
df_join
```

# Right Join

```python
df_transaction['CustomerId '].unique()
```

```python
df_join = pd.merge(df_left, df_right, how='right', on='CustomerId ')
df_join
```

# Inner Join

```python
df_join = pd.merge(df_left, df_right, how='inner', on='CustomerId ')
df_join
```

```python
df_join = pd.merge(df_left, df_right, how='outer', on='CustomerId ')
df_join
```
